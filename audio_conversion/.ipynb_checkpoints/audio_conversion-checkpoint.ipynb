# Audio Conversion
This jupyter notebook shows how t convert a wav file into a binary array.

```python
import wave
import struct
import sys

def wav_to_floats(wave_file):
    w = wave.open(wave_file)
    astr = w.readframes(w.getnframes())
    # convert binary chunks to short 
    a = struct.unpack("%ih" % (w.getnframes()* w.getnchannels()), astr)
    a = [float(val) / pow(2, 15) for val in a]
    return a

signal = wav_to_floats("InputSignal_f32_1kHz_15kHz'J.wav")
```

```python
import matplotlib.pyplot as plt
plt.plot(signal)
plt.show()
```

```python
import numpy as np
np.savetxt('audio.dat', signal)
```
